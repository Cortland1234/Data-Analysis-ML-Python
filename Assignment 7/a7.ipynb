````python
#-----------------------------
#Creating DataScientist class
#-----------------------------

def minkowskiDist(v1, v2, p):
    """Assumes v1 and v2 are equal-length arrays of numbers
       Returns Minkowski distance of order p between v1 and v2"""
    dist = 0.0
    for i in range(len(v1)):
        dist += abs(v1[i] - v2[i])**p
    return dist**(1/p)

def SortMembers(data):
    paid = 0
    unpaid = 0
    for d in data:
        if d[2] == 0:
            unpaid += 1
        else:
            paid += 1
    return paid, unpaid

class DataScientist(object):
    
    featureNames = ('Tenure', 'Salary', 'Paid')
    
    def __init__(self, tenure, salary, paid=0):
        self.tenure = tenure
        self.salary = salary
        self.paid = paid
        if paid == 0:
            self.label = 'Unpaid'
        else:
            self.label = 'Paid'
        self.features = [self.tenure, self.salary, self.paid]
    def getFeatures(self):
        return self.features
    def getLabel(self):
        return self.label
    def distance(self, other):
        return minkowskiDist(self.getFeatures(),
                             other.getFeatures(), 2)
    
    def __str__(self):
        return f'{self.label:<6}: Tenure={self.tenure:>4}, Salary={self.salary:>6}, Paid= {self.paid}'


def CreateScientists(data): #This function a list of scientists from the dataset
    scilist = []
    for d in data:
        member = DataScientist(d[0], d[1], d[2])
        scilist.append(member)
    return scilist

scilist = CreateScientists(data)
print(f'{len(scilist)} scientists in the list.')  
````

````python
#-------------------------------------
#Definitions of Metrics
#-------------------------------------
import random 

def accuracy(truePos, falsePos, trueNeg, falseNeg):
    numerator = truePos + trueNeg
    denominator = truePos + trueNeg + falsePos + falseNeg
    return numerator/denominator

def sensitivity(truePos, falseNeg):
    try:
        return truePos/(truePos + falseNeg)
    except ZeroDivisionError:
        return float('nan')
    
def specificity(trueNeg, falsePos):
    try:
        return trueNeg/(trueNeg + falsePos)
    except ZeroDivisionError:
        return float('nan')
    
def posPredVal(truePos, falsePos):
    try:
        return truePos/(truePos + falsePos)
    except ZeroDivisionError:
        return float('nan')
    
def negPredVal(trueNeg, falseNeg):
    try:
        return trueNeg/(trueNeg + falseNeg)
    except ZeroDivisionError:
        return float('nan')
    
def getStats(truePos, falsePos, trueNeg, falseNeg, toPrint = True):
    accur = accuracy(truePos, falsePos, trueNeg, falseNeg)
    sens = sensitivity(truePos, falseNeg)
    spec = specificity(trueNeg, falsePos)
    ppv = posPredVal(truePos, falsePos)
    if toPrint:
        print(' Accuracy =', round(accur, 3))
        print(' Sensitivity =', round(sens, 3))
        print(' Specificity =', round(spec, 3))
        print(' Pos. Pred. Val. =', round(ppv, 3))
    return (accur, sens, spec, ppv)

#-------------------------------
#Leave-Out Method and Splitting
#-------------------------------
import random

def leaveOneOut(data, method, toPrint = True):
    truePos, falsePos, trueNeg, falseNeg = 0, 0, 0, 0
    for i in range(len(data)):
        testCase = data[i]
        set = data[0:i] + data[i+1:]
        results = method(set, [testCase])
        truePos += results[0]
        falsePos += results[1]
        trueNeg += results[2]
        falseNeg += results[3]
    if toPrint:
        getStats(truePos, falsePos, trueNeg, falseNeg)
    return truePos, falsePos, trueNeg, falseNeg

def split80_20(data):
    sampleIndices = random.sample(range(len(data)),
                                  len(data)//5)
    set1, set2 = [], []
    for i in range(len(data)):
        if i in sampleIndices:
            set2.append(data[i])
        else:
            set1.append(data[i])
    return set1, set2

def randomSplits(data, method, numSplits, toPrint = True):
    truePos, falsePos, trueNeg, falseNeg = 0, 0, 0, 0
    random.seed(0)
    for t in range(numSplits):
        set1, set2 = split80_20(data)
        results = method(set1, set2)
        truePos += results[0]
        falsePos += results[1]
        trueNeg += results[2]
        falseNeg += results[3]
    getStats(truePos/numSplits, falsePos/numSplits,
             trueNeg/numSplits, falseNeg/numSplits, toPrint)
    return truePos/numSplits, falsePos/numSplits,\
             trueNeg/numSplits, falseNeg/numSplits

import sklearn.linear_model

def buildModel(scilist, toPrint = False):
    featureVecs, labels = [],[]
    for s in scilist:
        featureVecs.append(s.getFeatures())
        labels.append(s.getLabel())
    LogisticRegression = sklearn.linear_model.LogisticRegression
    model = LogisticRegression().fit(featureVecs, labels)
    
    
    if toPrint:
        obj = type(scilist[0])
        print('Model.classes_ =', model.classes_)
        for i in range(len(model.coef_)):
            print('For label', model.classes_[1])
            for j in range(len(model.coef_[0])):
                print('   ', obj.featureNames[j], '=',
                      model.coef_[0][j])
    return model

model = buildModel(scilist, toPrint=False)

def applyModel(model, testSet, label, prob = 0.75):
    FeatureVecs = [s.features for s in testSet]
    probs = model.predict_proba(FeatureVecs)
    truePos, falsePos, trueNeg, falseNeg = 0, 0, 0, 0
    ci = list(model.classes_).index(label)
    for i in range(len(probs)):
        if probs[i][ci] > prob:
            if testSet[i].getLabel() == label:
                truePos += 1
            else:
                falsePos += 1
        else:
            if testSet[i].getLabel() != label:
                trueNeg += 1
            else:
                falseNeg += 1
    return truePos, falsePos, trueNeg, falseNeg
#----------------------
#Logistic Regression
#----------------------

def lr(data1, data2, prob = 0.75, label='Unpaid'):
    model = buildModel(data1)
    return applyModel(model, data2, label, prob)


random.seed(0)
numSplits = 10
print('Average of', numSplits, '80/20 splits LR')
truePos1, falsePos1, trueNeg1, falseNeg1 =\
    randomSplits(scilist, lr, numSplits)

print('Average of LOO testing using LR')
truePos2, falsePos2, trueNeg2, falseNeg2 =\
    leaveOneOut(scilist, lr)

import sklearn

def BuildROC(set1, set2, title, label, plot=True):
    model = buildModel(set1, True)
    xVals, yVals = [], []
    p = 0.0
    while p <= 1.0:
        true_pos, false_pos, true_neg, false_neg =\
         applyModel(model, set2, label, p)
        xVals.append(1.0-specificity(true_neg, false_pos))
        yVals.append(sensitivity(true_pos, false_neg))
        p += 0.01
    auroc = sklearn.metrics.auc(xVals, yVals)
    if plot:
        plt.plot(xVals, yVals)
        plt.plot([0,1], [0,1])
        plt.title(f'{title}\nAUROC = {auroc:.4f}')
        plt.grid(True, alpha=0.5)
        plt.xlabel('1 - (specificity)')
        plt.ylabel('sensitivity')
        plt.show()
    return auroc

set1, set2 = split80_20(scilist)
n = 100
word = 'Unpaid'
auroc = BuildROC(set1, set2, f'ROC for predicting {word}, {n} Splits', word)
print(f'AUROC = {auroc:.4f}')
````

### Problem 4 KNN

In this problem, you need to write function(s) to build a classifier using KNN algorithm.  Then, you need to apply the test methods (leaveOneOut and randomSplit) to evaluate the learned classifer in terms of accuracy, sensitivity, specificity, and positive predicative value.

```

Average of LOO testing using KNN (k=3)
 Accuracy = 0.85

 Specificity = 0.932
 Pos. Pred. Val. = 0.762

 ```

````python
#-----------------------------
#Using KNN Algorithm
#-----------------------------

def findKNearest(dataset, set, k):
    kNearest, distances = [], []
    #Build lists containing first k examples and their distances
    for i in range(k):
        kNearest.append(set[i])
        distances.append(dataset.distance(set[i]))
    maxDist = max(distances) #Get maximum distance
    #Look at examples not yet considered
    for x in set[k:]:
        dist = dataset.distance(x)
        if dist < maxDist:
            #replace farther neighbor by this one
            maxIndex = distances.index(maxDist)
            kNearest[maxIndex] = x
            distances[maxIndex] = dist
            maxDist = max(distances)      
    return kNearest, distances

def KNearestClassify(set1, set2, label, k):
    truePos, falsePos, trueNeg, falseNeg = 0, 0, 0, 0
    for s in set2:
        nearest, distances = findKNearest(s, set1, k)
        #conduct vote
        numMatch = 0
        for i in range(len(nearest)):
            if nearest[i].getLabel() == label:
                numMatch += 1
        if numMatch > k//2: #guess label
            if s.getLabel() == label:
                truePos += 1
            else:
                falsePos += 1
        else: #guess not label
            if s.getLabel() != label:
                trueNeg += 1
            else:
                falseNeg += 1
    return truePos, falsePos, trueNeg, falseNeg

knn = lambda set1, set2:\
             KNearestClassify(set1, set2,
                              'Unpaid', 3)
numSplits = 10
print('Average of', numSplits,
      '80/20 splits using KNN (k=3)')
truePos, falsePos, trueNeg, falseNeg =\
      randomSplits(scilist, knn, numSplits)

print('Average of LOO testing using KNN (k=3)')
truePos, falsePos, trueNeg, falseNeg =\
      leaveOneOut(scilist, knn)
````

````python
#--------------------
#Getting csv data and creating Customer class
#--------------------

import csv

def getCustomers(filename):
    cusdict = {}
    with open(filename) as file:
        csv_reader = csv.reader(file, delimiter=',')
        n = -1
        for row in csv_reader:
            n += 1
            if n == 0:
                headers = row
            else:
                cusdict[row[0]] = row[1:]
    return headers, cusdict

headers, cusdict = getCustomers('loan_data.csv')
nnn = ''
for head in headers:
    nnn += head.upper()+', '

class Customer:
    featureNames = ('DTI', 'Score', 'Payment-Income Ratio')
    def __init__(self, id, outcome, dti, score, ratio):
        self.id = id
        self.outcome = outcome
        self.dti = float(dti)
        self.score = float(score)
        self.ratio = float(ratio)
        self.features = (outcome, dti, score, ratio)
        if outcome == 'paid off':
            self.label = 'Paid Off'
        else:
            self.label = 'Default'
        self.features = (self.dti, self.score, self.ratio)
        
    def getFeatures(self):
        return self.features
    
    def getLabel(self):
        return self.label  
    
    def distance(self, other):
        return minkowskiDist(self.getFeatures(),
                             other.getFeatures(), 2)  
    def __str__(self):
        return f'{self.id:<5}: Outcome={self.outcome:<8}, DTI={self.dti:<5}, Score={self.score:.2f}, Ratio={self.ratio:.6f}'
    
def CreateCustomers(data):
    cuslist = []
    for t in data.keys():
        v = data.get(t)
        person = Customer(t, v[0], v[1], v[2], v[3])
        cuslist.append(person)
    return cuslist

cuslist = CreateCustomers(cusdict)
print(f'{len(cuslist)} customers in the list.')
        
````
