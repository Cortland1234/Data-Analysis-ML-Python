```python
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#--------------------------------------------------------------------------
#Create classes for clusters
#--------------------------------------------------------------------------

def minkowskiDist(v1, v2, p):
    #Assumes v1 and v2 are equal length arrays of numbers
    dist = 0
    for i in range(len(v1)):
        dist += abs(v1[i] - v2[i])**p
    return dist**(1/p)

class Student(object):   
    def __init__(self, name, loc):
        self.name = name
        self.loc = loc
        
    def dimensionality(self):
        return int(len(self.loc))
    
    def get_location(self):
        return self.loc
    
    def getName(self):
        return self.name
    
    def distance(self, other):
        return minkowskiDist(self.loc, other.get_location(), 2)
    
    def __str__(self):
        return self.name +' '+self.loc

students = []
n = 0
for i in inputs:
    student = Student(namelist[n], i)
    students.append(student)
    n += 1
    
class Cluster(object):
    def __init__(self, students):
        """Assumes examples a non-empty list of Examples"""
        self.students = students
        self.centroid = self.computeCentroid()
        
    def update(self, students):
        """Assume examples is a non-empty list of Examples
           Replace examples; return amount centroid has changed"""
        oldCentroid = self.centroid
        self.students = students
        self.centroid = self.computeCentroid()
        return oldCentroid.distance(self.centroid)
    def computeCentroid(self):
        vals = np.array([0.0]*self.students[0].dimensionality())
        for e in self.students: #compute mean
            vals += e.get_location()
        centroid = Student('centroid', vals/len(self.students))
        return centroid
    def getCentroid(self):
        return self.centroid
    def variability(self):
        totDist = 0
        for e in self.students:
            totDist += (e.distance(self.centroid))**2
        return totDist      
    def members(self):
        for e in self.students:
            yield e
    def __str__(self):
        names = []
        for e in self.students:
            names.append(e.getName())
        names.sort()
        result = 'Cluster with centroid '\
               + str(self.centroid.getFeatures()) + ' contains:\n  '
        for e in names:
            result = result + e + ', '
        return result[:-2] #remove trailing comma and space  

    
#---------------------------------------------------------------------
#Calculate dissimilarity and define k-means
#---------------------------------------------------------------------
def dissimilarity(clusters):
    """Assumes clusters a list of clusters
       Returns a measure of the total dissimilarity of the
       clusters in the list"""
    totDist = 0
    for c in clusters:
        totDist += c.variability()
    return totDist

def kmeans(students, k, verbose = False):
    
    #Get k randomly chosen initial centroids, create cluster for each
    initialCentroids = random.sample(students, k)
    
    clusters = []
    for e in initialCentroids:
        clusters.append(Cluster([e]))   
        
    #Iterate until centroids do not change
    converged = False
    numIterations = 0
    
    while not converged:
        
        numIterations += 1
        
        #Create a list containing k distinct empty lists
        newClusters = []
        for i in range(k):
            newClusters.append([]) 
            
        #Associate each example with closest centroid
        for e in students:
            #Find the centroid closest to e
            smallestDistance = e.distance(clusters[0].getCentroid())
            index = 0
            for i in range(1, k):
                distance = e.distance(clusters[i].getCentroid())
                if distance < smallestDistance:
                    smallestDistance = distance
                    index = i
            #Add e to the list of examples for appropriate cluster
            newClusters[index].append(e)
            
        for c in newClusters: #Avoid having empty clusters
            if len(c) == 0:
                raise ValueError('Empty Cluster')
                
        #Update each cluster; check if a centroid has changed
        converged = True
        for i in range(k):
            if clusters[i].update(newClusters[i]) > 0.0:
                converged = False
                
        if verbose:
            print('Iteration #' + str(numIterations))
            for c in clusters:
                print(c)
            print('') #add blank line
            fig = plt.figure()
            fig.set_size_inches(12, 8)
            plotClustering(clusters)
            plt.show()
            
    return clusters

def trykmeans(students, numClusters, numTrials, verbose = False):
    """Calls kmeans numTrials times and returns the result with the
          lowest dissimilarity"""
    
    if verbose:
        print("-----Trial 0:----")
    best = None
    while (best == None):
        try:
            best = kmeans(students, numClusters, verbose)
        except ValueError:
            continue #If failed, try again
    
    minDissimilarity = dissimilarity(best)
    
    trial = 1
    while trial < numTrials:
        if verbose:
            print("-----Trial " + str(trial) +":----")
            
        try:
            clusters = kmeans(students, numClusters, verbose)
        except ValueError:
            continue #If failed, try again
            
        currDissimilarity = dissimilarity(clusters)
        if currDissimilarity < minDissimilarity:
            best = clusters
            minDissimilarity = currDissimilarity
        trial += 1
    return best
```

## Problem 4: Plotting the Customers
For this problem, you need to plot the customers so that their *spending scores* and *annual incomes* could be visualized easily.

```python
#---------------------------------
#Import csv data
#---------------------------------

import csv

csvdata = []
f = 'shoppingdata.csv'
with open(f) as csvf:
    rows = csv.reader(csvf, delimiter = ',')
    for r in rows:
        try:
            c = str(r[0])
            gender = str(r[1][0])
            age = int(r[2])
            aninc = int(r[3])
            score = int(r[4])
            cust = [c, gender, age, aninc, score]
            csvdata.append(cust)
        except:
            pass
```

```python
#----------------------------------
#Create Shopper class and create list of Shoppers
#-----------------------------------

class Shopper(Student):
    def __init__(self, name, loc, gender, age):
        self.name = name
        self.loc = loc
        self.gender = gender
        self.age = age
        
    def getGender(self):
        return self.gender
    def getAge(self):
        return self.age
    
students = []
n = 0 
for c in csvdata:
    name, gender, age = c[0], c[1], c[2]
    xv, yv = c[4], c[3]
    student = Shopper(c[0], (xv,yv), gender, age)
    students.append(student)
    n += 1
```
